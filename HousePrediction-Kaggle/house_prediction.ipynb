```python
import pandas as pd
import numpy as np
from xgboost import XGBRegressor
```

```python
train = pd.DataFrame(pd.read_csv('train.csv')).drop('SalePrice', axis=1)
y = pd.DataFrame(pd.read_csv('train.csv'))['SalePrice']
test = pd.DataFrame(pd.read_csv('test.csv'))
sample = pd.DataFrame(pd.read_csv('sample_submission.csv'))
```

```python
train.head()
```

```python
test.head()
```

```python
sample.head()
```

```python
train = pd.get_dummies(train)
train.head()
```

```python
test = pd.get_dummies(test)
test.head()
```

```python
train, test = train.align(test, join='left', axis=1)
train.head()
```

```python
test.head()
```

```python
model = XGBRegressor()
```

```python
model.fit(train, y)
```

```python
prediction = model.predict(test)
```

```python
prediction
```

```python
indices = np.arange(1461, 2920)
put = pd.DataFrame({'Id':indices, 'SalePrice':prediction})
put.to_csv('prediction2.csv', mode='a', index=False)
```
